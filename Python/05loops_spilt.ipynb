# While loop

```python
x=1
while(x<10):
    print('value is',x)
    x=x+1
print("Done")
```

```python

x=1
while(x<10):
    if(x==7):
        break
    print('value is',x)
    x=x+1
    
print("Done")
```

```python
i=0
while (i<10):
    
    i=i+1
    if(i==5):
        continue
    print('value of i is',i)
print("Done")
```

```python
i=0;
while i <9:
    i=i+1
    if(i==3):
        continue
    print(i)
print("loop done")
```

```python
my_list=['apple','banana','orange','mango','grapes']
size=len(my_list)
size
```

```python
i=0
while i<size:
    if my_list[i]=='mango':
        print("mango found")
        break;
    i=i+1
else:
    print("mango not found")
```
